# Pandas

Pandas is an open-source data manipulation and analysis library for Python. It provides easy-to-use data structures, such as Series and DataFrame, along with a variety of functions to manipulate and analyze structured data. Pandas is widely used in data science, machine learning, and data analysis for tasks such as cleaning, transforming, and analyzing tabular data.

### Grouping and Aggregating

```python
data["SalaryUSD"].median()
```

```python
data["Hobbyist"].value_counts()
```

```python
data["SocialMedia"].value_counts(normalize=True)
```

```python
data["Country"].value_counts()
```

```python
country_group = data.groupby(["Country"])
```

```python
country_group.get_group("India")
```

```python
filt = data["Country"] == "United States"
data.loc[filt]["SocialMedia"].value_counts()
```

```python
filt = data["Country"] == "India"
data.loc[filt]["SocialMedia"].value_counts()
```

```python
country_group["SocialMedia"].value_counts(normalize=True).loc["China"]
```

```python
country_group["SocialMedia"].value_counts(normalize=True).loc["Russian Federation"]
```

```python
country_group["SalaryUSD"].median()
```

```python
country_group["SalaryUSD"].median().loc["Germany"]
```

```python
country_group["SalaryUSD"].agg(["median", "mean"])
```

```python
country_group["SalaryUSD"].agg(["median", "mean"]).loc["Canada"]
```

```python
filt = data["Country"] == "India"
data.loc[filt]["LanguageWorkedWith"].str.contains("Python").sum()
```

```python
country_group["LanguageWorkedWith"].apply(lambda x: x.str.contains("Python").sum())
```

```python
country_respondents = data["Country"].value_counts()
country_respondents
```

```python
country_uses_python = country_group["LanguageWorkedWith"].apply(lambda x: x.str.contains("Python").sum())
country_uses_python
```

```python
python_data = pd.concat([country_respondents, country_uses_python], axis="columns", sort=False)
python_data
```

```python
python_data["Percentage"] = (python_data["LanguageWorkedWith"]/data["Country"]) * 100
python_data
```

```python
python_data.sort_values(by="Percentage", ascending=False, inplace=True)
```

```python
python_data.head(50)
```
