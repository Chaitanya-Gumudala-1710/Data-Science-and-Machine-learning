# Scatter Plot

A scatter plot is a type of chart that displays individual data points on a two-dimensional graph. Each point represents the values of two variables, and the position of the point is determined by the values of these variables. Scatter plots are useful for visualizing the relationship and distribution of data points, particularly in identifying trends or patterns.

## Importing Required Libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

### Creating the data

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

### Scatter plot in python

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.scatter(x, y,s=100, c="green", marker="X", edgecolor="black")
plt.show()
```

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
```

### Plot data points with different sizes and colors, and includes a colorbar indicating the level of satisfaction.

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.scatter(x, y, s=sizes, c=colors, cmap="Greens", edgecolor="black", linewidth=1, alpha=0.75)
plt.colorbar().set_label("Satisfaction")
plt.show()
```

### Loading the csv data

```python
data = pd.read_csv("data.csv")
data
```

### Scatter plot of YouTube video data, visualizing the relationship between view count and likes.

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.scatter(data["view_count"], data["likes"], s=100, c=data["ratio"], cmap="summer", edgecolor="black", linewidth=1, alpha=0.75)
plt.xscale("log")
plt.yscale("log")
plt.colorbar().set_label("Like/Dislike Ratio")
plt.title("Trending YouTube Videos")
plt.xlabel("View Count")
plt.ylabel("Total Likes")
plt.show()
```
