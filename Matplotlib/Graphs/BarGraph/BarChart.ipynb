```python
with open("data.csv") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    language_counter = Counter()
    for row in csv_reader:
        language_counter.update(row["LanguagesWorkedWith"].split(";"))
```

```python
languages = []
popularity = []
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.bar(languages, popularity, color="#775FEC")
plt.xlabel("Programming Languages")
plt.ylabel("Number of People")
plt.title("Most Popular Languages")
plt.show()
```

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.barh(languages, popularity, color="#FF6666")
plt.xlabel("Number of People")
plt.title("Most Popular Languages")
plt.show()
```

```python
data = pd.read_csv("data.csv")
identity = data["Responder_id"]
programming_languages_responses = data["LanguagesWorkedWith"]
for response in programming_languages_responses:
    language_counter.update(response.split(";"))
```

```python
languages = []
popularity = []
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
languages.reverse()
popularity.reverse()
```

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.barh(languages, popularity, color="#6EB7CD")
plt.xlabel("Number of People")
plt.title("Most Popular Languages")
plt.show()
```
