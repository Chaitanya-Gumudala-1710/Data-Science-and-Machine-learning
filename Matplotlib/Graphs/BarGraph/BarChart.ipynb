# Bar Chart

A bar chart is a graphical representation of data where rectangular bars of varying lengths are used to represent different categories or groups. The length of each bar corresponds to the quantity or value associated with that category. Bar charts are effective for visualizing and comparing the distribution or relationship of data across different categories.

### Reading data from a CSV file and counts the occurrences of different programming languages.

```python
with open("data.csv") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    language_counter = Counter()
    for row in csv_reader:
        language_counter.update(row["LanguagesWorkedWith"].split(";"))
```

### Retrieves the top 15 languages and their popularity.

```python
languages = []
popularity = []
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
```

### Bar chart of programming language popularity.

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.bar(languages, popularity, color="#775FEC")
plt.xlabel("Programming Languages")
plt.ylabel("Number of People")
plt.title("Most Popular Languages")
plt.show()
```

### Horizontal Bar chart of programming language popularity.

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.barh(languages, popularity, color="#FF6666")
plt.xlabel("Number of People")
plt.title("Most Popular Languages")
plt.show()
```

### Reading data from a CSV file and counts programming language responses.

```python
data = pd.read_csv("data.csv")
identity = data["Responder_id"]
programming_languages_responses = data["LanguagesWorkedWith"]
for response in programming_languages_responses:
    language_counter.update(response.split(";"))
```

### Creating two lists, languages and popularity, by extracting data from the language_counter dictionary and reversing the order of the elements.

```python
languages = []
popularity = []
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
languages.reverse()
popularity.reverse()
```

### Horizontal bar chart to visualize the popularity of different programming languages.

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.barh(languages, popularity, color="#6EB7CD")
plt.xlabel("Number of People")
plt.title("Most Popular Languages")
plt.show()
```
