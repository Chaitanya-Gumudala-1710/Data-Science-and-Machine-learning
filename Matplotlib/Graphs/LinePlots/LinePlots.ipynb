```python
data = pd.read_csv("data.csv")
```

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.plot(data["Age"], data["All_Devs"], color="#444444", linestyle="--", label="All Developers")
plt.plot(data["Age"], data["Python"], label="Python")
plt.fill_between(data["Age"], data["Python"], data["All_Devs"].median(), 
                 where=(data["Python"] > data["All_Devs"].median()), 
                 interpolate=True, alpha=0.25)
plt.xlabel("Age")
plt.ylabel("Median Salary($)")
plt.title("Median Salary(USD) by age")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.plot(data["Age"], data["All_Devs"], color="#444444", linestyle="--", label="All Developers")
plt.plot(data["Age"], data["Python"], label="Python")
plt.fill_between(data["Age"], data["Python"], data["All_Devs"].median(), 
                 where=(data["Python"] > data["All_Devs"].median()), 
                 interpolate=True, alpha=0.25)
plt.fill_between(data["Age"], data["Python"], data["All_Devs"].median(), 
                 where=(data["Python"] <= data["All_Devs"].median()), 
                 interpolate=True, color="red", alpha=0.25)
plt.xlabel("Age")
plt.ylabel("Median Salary($)")
plt.title("Median Salary(USD) by age")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.plot(data["Age"], data["All_Devs"], color="#444444", linestyle="--", label="All Developers")
plt.plot(data["Age"], data["Python"], label="Python")
plt.fill_between(data["Age"], data["Python"], data["All_Devs"], 
                 where=(data["Python"] > data["All_Devs"]), 
                 interpolate=True, alpha=0.25, label="Above Average")
plt.fill_between(data["Age"], data["Python"], data["All_Devs"], 
                 where=(data["Python"] <= data["All_Devs"]), 
                 interpolate=True, color="red", alpha=0.25, label="Below Average")
plt.xlabel("Age")
plt.ylabel("Median Salary($)")
plt.title("Median Salary(USD) by age")
plt.legend()
plt.show()
```
