# Histogram

A histogram is a graphical representation of the distribution of a dataset. It is used to visualize the underlying frequency distribution of a set of continuous or discrete data. In a histogram, the data is divided into intervals (or bins), and the vertical bars represent the frequency or count of data points falling within each interval.

## Loading Required Libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

### Reading the data

```python
data = pd.read_csv("data.csv")
data
```

### Histogram plot of the "Age" column from a dataset.

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.hist(data["Age"], bins=50, edgecolor="black")
plt.title("Age of Respondents")
plt.xlabel("Age")
plt.ylabel("Total Respondents")
plt.axvline(data["Age"].median(), color="#fc4f30", label="Median Age")
plt.legend()
plt.show()
```
