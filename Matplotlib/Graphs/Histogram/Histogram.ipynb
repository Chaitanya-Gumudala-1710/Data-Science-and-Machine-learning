# Histogram

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data
```

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.hist(data["Age"], bins=50, edgecolor="black")
plt.title("Age of Respondents")
plt.xlabel("Age")
plt.ylabel("Total Respondents")
plt.axvline(data["Age"].median(), color="#fc4f30", label="Median Age")
plt.legend()
plt.show()
```
