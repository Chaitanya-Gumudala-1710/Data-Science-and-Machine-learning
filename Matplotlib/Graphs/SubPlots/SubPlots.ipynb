# Sub Plots

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.plot(data["Age"], data["Python"], label='Python')
plt.plot(data["Age"], data["JavaScript"], label='JavaScript')
plt.plot(data["Age"], data["All_Devs"], color='#444444', linestyle='--', label='All Devs')
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize = (25, 25), sharex=True)
plt.xkcd()
ax1.plot(data["Age"], data["Python"], label='Python')
ax2.plot(data["Age"], data["JavaScript"], label='JavaScript')
ax2.plot(data["Age"], data["All_Devs"], color='#444444', linestyle='--', label='All Devs')
ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')
ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
fig1, ax1 = plt.subplots(figsize = (25, 25))
fig2, ax2 = plt.subplots(figsize = (25, 25))
plt.xkcd()
ax1.plot(data["Age"], data["Python"], label='Python')
ax2.plot(data["Age"], data["JavaScript"], label='JavaScript')
ax2.plot(data["Age"], data["All_Devs"], color='#444444', linestyle='--', label='All Devs')
ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')
ax2.legend()
ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```
