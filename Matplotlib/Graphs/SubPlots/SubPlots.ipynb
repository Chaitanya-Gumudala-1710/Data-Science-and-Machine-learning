# Sub Plots

## Loading required libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

### Loading the data

```python
data = pd.read_csv("data.csv")
data.head()
```

### Line plot with multiple lines to compare the median salary of Python, JavaScript, and all developers based on age.

```python
plt.figure(figsize=(20, 10), facecolor="White")
plt.xkcd()
plt.plot(data["Age"].values, data["Python"].values, label='Python')
plt.plot(data["Age"].values, data["JavaScript"].values, label='JavaScript')
plt.plot(data["Age"].values, data["All_Devs"].values, color='#444444', linestyle='--', label='All Devs')
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

### Creating a figure with two subplots and plots data on each subplot using matplotlib.

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize = (25, 25), sharex=True)
plt.xkcd()
ax1.plot(data["Age"].values, data["Python"].values, label='Python')
ax2.plot(data["Age"].values, data["JavaScript"].values, label='JavaScript')
ax2.plot(data["Age"].values, data["All_Devs"].values, color='#444444', linestyle='--', label='All Devs')
ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')
ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

###  Creating two subplots using matplotlib to visualize median salaries for different programming languages by age.

```python
fig1, ax1 = plt.subplots(figsize = (25, 25))
fig2, ax2 = plt.subplots(figsize = (25, 25))
plt.xkcd()
ax1.plot(data["Age"].values, data["Python"].values, label='Python')
ax2.plot(data["Age"].values, data["JavaScript"].values, label='JavaScript')
ax2.plot(data["Age"].values, data["All_Devs"].values, color='#444444', linestyle='--', label='All Devs')
ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')
ax2.legend()
ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```
