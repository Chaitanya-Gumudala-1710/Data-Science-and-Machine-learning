# Scatter Plot

A scatter plot is a type of chart that displays individual data points on a two-dimensional graph. Each point represents the values of two variables, and the position of the point is determined by the values of these variables. Scatter plots are useful for visualizing the relationship and distribution of data points, particularly in identifying trends or patterns.

## Loading the required libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the dataset

```python
data = sns.load_dataset("penguins")
```

### Plotting the scatter plot

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex", size="sex", sizes=(50, 25))
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", 
                style="sex", size="sex", sizes=(100, 50), palette="twilight_shifted")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", 
                style="sex", size="sex", sizes=(100, 50), palette="PiYG", alpha=1)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", 
                style="sex", size="sex", sizes=(100, 100), palette="turbo", alpha=1, markers={"Male":"*", "Female":"o"})
plt.show()
```
