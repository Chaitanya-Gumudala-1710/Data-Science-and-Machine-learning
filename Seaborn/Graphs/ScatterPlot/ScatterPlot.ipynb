# Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("penguins")
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex", size="sex", sizes=(50, 25))
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", 
                style="sex", size="sex", sizes=(100, 50), palette="twilight_shifted")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", 
                style="sex", size="sex", sizes=(100, 50), palette="PiYG", alpha=1)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", 
                style="sex", size="sex", sizes=(100, 100), palette="turbo", alpha=1, markers={"Male":"*", "Female":"o"})
plt.show()
```
