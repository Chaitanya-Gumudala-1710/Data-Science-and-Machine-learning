# Pair Plot

A pair plot, also known as a scatterplot matrix, is a grid of scatterplots where each variable is compared to every other variable in a dataset. Pair plots are useful for visualizing relationships and distributions between pairs of variables in a multivariate dataset.

## Loading the required libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the dataset

```python
data = sns.load_dataset("tips")
data
```

### Plotting a pairplot

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data)
plt.show()
```

### Plotting a pair plot by differenciating "sex"

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, hue="sex")
plt.show()
```

### Plotting plair plot of total_bill vs tips with "sex" be differeniciating factor

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, vars=["total_bill", "tip"], hue="sex")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, vars=["total_bill", "tip"], hue="sex", hue_order=["Female", "Male"])
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, vars=["total_bill", "tip"], hue="sex", hue_order=["Female", "Male"], palette="YlOrRd")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, x_vars=["total_bill", "tip"], hue="sex", hue_order=["Female", "Male"], palette="spring")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, hue="sex", hue_order=["Female", "Male"], palette="hsv", kind="reg")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, hue="sex", hue_order=["Female", "Male"], palette="hsv", kind="kde")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, hue="sex", hue_order=["Female", "Male"], palette="nipy_spectral", kind="kde", diag_kind="hist")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, hue="sex", hue_order=["Female", "Male"], palette="Accent", markers=["*", ">"])
plt.show()
```
