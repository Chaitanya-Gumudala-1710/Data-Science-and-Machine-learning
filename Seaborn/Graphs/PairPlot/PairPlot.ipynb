# Pair Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("tips")
data
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data)
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, hue="sex")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, vars=["total_bill", "tip"], hue="sex")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, vars=["total_bill", "tip"], hue="sex", hue_order=["Female", "Male"])
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, vars=["total_bill", "tip"], hue="sex", hue_order=["Female", "Male"], palette="YlOrRd")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, x_vars=["total_bill", "tip"], hue="sex", hue_order=["Female", "Male"], palette="spring")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, hue="sex", hue_order=["Female", "Male"], palette="hsv", kind="reg")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, hue="sex", hue_order=["Female", "Male"], palette="hsv", kind="kde")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, hue="sex", hue_order=["Female", "Male"], palette="nipy_spectral", kind="kde", diag_kind="hist")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(data, hue="sex", hue_order=["Female", "Male"], palette="Accent", markers=["*", ">"])
plt.show()
```
