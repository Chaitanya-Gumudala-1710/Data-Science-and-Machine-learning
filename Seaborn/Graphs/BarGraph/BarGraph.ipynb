# Bar Graph

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("penguins")
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.island, y=data.bill_length_mm)
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.island, y=data.bill_length_mm, hue=data.sex)
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.island, y=data.bill_length_mm, hue=data.sex, order=["Dream", "Torgersen", "Biscoe"])
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.island, y=data.bill_length_mm, hue=data.sex, 
            order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"])
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.island, y=data.bill_length_mm, hue=data.sex, 
            order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], ci=50)
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.island, y=data.bill_length_mm, hue=data.sex, 
            order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], ci=25, n_boot=3)
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.bill_length_mm, y=data.island, hue=data.sex, 
            order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], orient="h")
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.bill_length_mm, y=data.island, hue=data.sex, 
            order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], orient="h", color="r")
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.bill_length_mm, y=data.island, hue=data.sex, 
            order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], orient="h", palette="icefire")
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.bill_length_mm, y=data.island, hue=data.sex, 
            order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], orient="h", palette="ocean", errcolor="g", errwidth=10)
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.island, y=data.bill_length_mm, hue=data.sex, order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], orient="v", palette="ocean", errcolor="g", errwidth=5, capsize=0.25)
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.set(style="darkgrid")
sns.barplot(x=data.island, y=data.bill_length_mm, hue=data.sex, order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], orient="v", palette="gist_rainbow", errcolor="g", errwidth=5, alpha=0.5)
plt.show()
```
