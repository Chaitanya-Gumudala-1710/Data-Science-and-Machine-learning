# Bar Graph

A bar chart is a graphical representation of data where rectangular bars of varying lengths are used to represent different categories or groups. The length of each bar corresponds to the quantity or value associated with that category. Bar charts are effective for visualizing and comparing the distribution or relationship of data across different categories.

### Loading required libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the penguin dataset

```python
data = sns.load_dataset("penguins")
```

### Bar graph to visualize the relationship between the "island" and "bill_length_mm"

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.island, y=data.bill_length_mm)
plt.show()
```

### Bar graph to visualize the relationship between the "island" vs "bill_length_mm" and differentiates the bars by the 'sex'

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.island, y=data.bill_length_mm, hue=data.sex)
plt.show()
```

### Giving order to plot the graph

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.island, y=data.bill_length_mm, hue=data.sex, order=["Dream", "Torgersen", "Biscoe"])
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.island, y=data.bill_length_mm, hue=data.sex, 
            order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"])
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.island, y=data.bill_length_mm, hue=data.sex, 
            order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], errorbar=('ci', 50))
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.island, y=data.bill_length_mm, hue=data.sex, 
            order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], errorbar=('ci', 25), n_boot=3)
plt.show()
```

### Plotting horizontal bar graph

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.bill_length_mm, y=data.island, hue=data.sex, 
            order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], orient="h")
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.bill_length_mm, y=data.island, hue=data.sex, 
            order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], orient="h", palette='dark:r')
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(x=data.bill_length_mm, y=data.island, hue=data.sex, 
            order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], orient="h", palette="icefire")
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.set_theme(style="darkgrid")
sns.barplot(x=data.island, y=data.bill_length_mm, hue=data.sex, order=["Dream", "Torgersen", "Biscoe"], 
            hue_order=["Female", "Male"], orient="v", palette="gist_rainbow", err_kws={'color': 'g', 'linewidth': 5}, alpha=0.5)
plt.show()
```
