# Strip Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("tips")
data
```

```python
plt.figure(figsize=(20, 10))
sns.stripplot(x="day", y="total_bill", data=data)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.stripplot(x="day", y="total_bill", data=data, hue="sex")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.stripplot(x="day", y="total_bill", data=data, hue="sex", palette="jet_r")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.stripplot(x="day", y="total_bill", data=data, hue="sex", palette="rocket", linewidth=0.75)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.stripplot(x="day", y="total_bill", data=data, hue="sex", palette="rocket", linewidth=0.75, jitter=1)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.stripplot(x="day", y="total_bill", data=data, hue="sex", palette="flag", linewidth=0.75, size=7.5)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.stripplot(x="day", y="total_bill", data=data, hue="sex", palette="flag", linewidth=0.75, size=7.5, marker="*")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.stripplot(x="total_bill", data=data, marker="*")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.stripplot(y="total_bill", data=data, marker="*")
plt.show()
```
