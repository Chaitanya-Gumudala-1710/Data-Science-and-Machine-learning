# Histogram

A histogram is a graphical representation of the distribution of a dataset. It is used to visualize the underlying frequency distribution of a set of continuous or discrete data. In a histogram, the data is divided into intervals (or bins), and the vertical bars represent the frequency or count of data points falling within each interval.

## Loading the required libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the dataset

```python
data = sns.load_dataset("penguins")
```

### Plotting the historgam of bill_length column

```python
sns.displot(data["bill_length_mm"])
plt.show()
```

### Plotting the historgam of flipper_length column

```python
sns.displot(data["flipper_length_mm"])
plt.show()
```

### Plotting the historgam of flipper length column by specifying the bins

```python
sns.displot(data["flipper_length_mm"], bins=[170, 180, 190, 200, 210, 220, 230, 240])
plt.show()
```
