# Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("penguins")
```

```python
sns.displot(data["bill_length_mm"])
plt.show()
```

```python
sns.displot(data["flipper_length_mm"])
plt.show()
```

```python
sns.displot(data["flipper_length_mm"], bins=[170, 180, 190, 200, 210, 220, 230, 240])
plt.show()
```

```python
sns.displot(data["flipper_length_mm"], bins=[170, 180, 190, 200, 210, 220, 230, 240], kde=True)
plt.show()
```

```python
sns.displot(data["flipper_length_mm"], bins=[170, 180, 190, 200, 210, 220, 230, 240], kde=True, rug=True)
plt.show()
```

```python
sns.displot(data["flipper_length_mm"], bins=[170, 180, 190, 200, 210, 220, 230, 240], kde=True, rug=True, color="r")
plt.show()
```
