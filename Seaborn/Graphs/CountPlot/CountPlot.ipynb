# Count Plot

A count plot is a type of categorical plot that shows the count of observations in each category or bin

## Loading the required libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the data

```python
data = sns.load_dataset("tips")
data
```

### Creating a count plot of the "sex" variable

```python
plt.figure(figsize=(20, 10))
sns.countplot(x="sex", data=data)
plt.show()
```

### Creating a count plot of the "sex" variable with hue as "smoker"

```python
plt.figure(figsize=(20, 10))
sns.countplot(x="sex", data=data, hue="smoker")
plt.show()
```

### Creating a horizontal count plot of the "sex" variable with hue as "smoker"

```python
plt.figure(figsize=(20, 10))
sns.countplot(y="sex", data=data, hue="smoker")
plt.show()
```

### Changing the colours of the count plots

```python
plt.figure(figsize=(20, 10))
sns.countplot(x="sex", data=data, hue="smoker", palette="BuGn")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x="sex", data=data, hue="smoker", palette="magma", saturation=0.5)
plt.show()
```
