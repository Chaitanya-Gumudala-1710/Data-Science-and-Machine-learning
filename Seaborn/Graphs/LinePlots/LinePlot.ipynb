# Line Plot

A line plot is a type of chart that displays data points over a continuous interval or time span and connects them with straight line segments. Line plots are commonly used to visualize trends, patterns, or relationships in data.

## Loading the required libraries

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

### Loading the dataset

```python
data = sns.load_dataset("penguins")
```

```python
data
```

### Plotting line plot

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="flipper_length_mm", data=data)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex", palette="Accent")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex", palette="Accent", markers=["o", ">"])
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex", 
             palette="Accent", markers=["o", ">"], dashes=False)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex", 
             palette="Accent", markers=["o", ">"], dashes=False, legend=False)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex", 
             palette="Accent", markers=["o", ">"], dashes=False, legend="full")
plt.grid()
plt.show()
```
