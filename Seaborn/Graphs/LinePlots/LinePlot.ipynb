# Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("penguins")
```

```python
data
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="flipper_length_mm", data=data)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex", palette="Accent")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex", palette="Accent", markers=["o", ">"])
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex", 
             palette="Accent", markers=["o", ">"], dashes=False)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex", 
             palette="Accent", markers=["o", ">"], dashes=False, legend=False)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue="sex", style="sex", 
             palette="Accent", markers=["o", ">"], dashes=False, legend="full")
plt.grid()
plt.show()
```
