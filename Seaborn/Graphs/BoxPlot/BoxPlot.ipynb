# Box Plot

Box plots, also known as box-and-whisker plots, are a graphical representation of the distribution of a dataset. They provide a concise summary of the central tendency, spread, and shape of the distribution.

## Loading the required libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the dataset

```python
data = sns.load_dataset("tips")
data
```

### Plotting the box plot day vs total bill

```python
plt.figure(figsize=(20, 10))
sns.set_theme(style="whitegrid")
sns.boxplot(x=data["day"], y=data["total_bill"])
plt.show()
```

### Plotting the horizontal box plot totalbill vs day

```python
plt.figure(figsize=(20, 10))
sns.set_theme(style="whitegrid")
sns.boxplot(x=data["total_bill"], y=data["day"])
plt.show()
```

### Box plot to visualize the distribution of "total_bill" values based on the "day" and "sex" variables.

```python
plt.figure(figsize=(20, 10))
sns.set_theme(style="whitegrid")
sns.boxplot(x=data["day"], y=data["total_bill"], hue=data["sex"])
plt.show()
```

### Box plot visualization for the "total_bill" column of a dataset 

```python
plt.figure(figsize=(20, 10))
sns.set_theme(style="whitegrid")
sns.boxplot(y=data["total_bill"])
plt.show()
```
